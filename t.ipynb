```python
import pandas as pd
import os

epmain = pd.read_csv(
    os.path.join("D:", "datasets", "easyportraitv2", "easyportraitv2_main.csv")
)
wfmain = pd.read_csv(os.path.join("D:", "datasets", "widerface", "widerface_main.csv"))
```

```python
wfmain["source"] = "widerface"
wfmain
```

```python
def compute_abs_path(row: pd.Series) -> str:
    relative_path = row["image"]
    subset = row["subset"]
    abs_path = os.path.join(
        "D:\\",
        "datasets",
        "widerface",
        f"WIDER_{subset}",
        f"WIDER_{subset}",
        "images",
        relative_path,
    )
    return abs_path


wfmain["image"] = wfmain.apply(compute_abs_path, axis=1)
wfmain
```

```python
wfmain["image"].apply(lambda x: os.path.exists(x)).all()
```

```python
main = pd.concat([epmain, wfmain]).sample(frac=1).reset_index(drop=True)
```

```python
main.to_csv(os.path.join("D:", "datasets", "main.csv"), index=False)
```
